```python

def get_sum(df, image_name, listed_feature):
    sum = 0
    for i in range(1, 5):
        sheet_name = f"Sheet{i}"
        image_col = f"{sheet_name}.image_name"
        feature_col = f"{sheet_name}.listed_features_{i}.x"
        n_col = f"{sheet_name}.n"
        sum += df[(df[image_col] == image_name) & (df[feature_col] == listed_feature)][n_col].sum()
    return sum
```
